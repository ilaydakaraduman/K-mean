```python
#QUESTON4-!!!!!!!!! My 5th Question under Question 4  --->QUESTON5
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
class KMeans:
    def __init__(self, K=3, max_iters=100, plot_steps=False):
        self.K = K
        self.max_iters = max_iters # it keeps max iterations counts
        self.plot_steps = plot_steps

        self.clusters = [[] for _ in range(self.K)] # default initialization
        self.centroids = []  # default center initialization
    isLast = False
    def predict(self, X): # main cod
        self.X = X# flowers
        self.n_samples, self.n_features = X.shape

        # initialize
        random_sample_idxs = np.random.choice(self.n_samples, self.K, replace=False)
        self.centroids = [self.X[idx] for idx in random_sample_idxs] # it chooses random center from flowers
        # Optimize clusters
        for _ in range(self.max_iters):
            # Assign samples to closest centroids (create clusters)
            self.clusters = self._create_clusters(self.centroids) #flowers are added to their current set

            if self.plot_steps:
                self.plot()

            # Calculate new centroids from the clusters
            centroids_old = self.centroids
            self.centroids = self._get_centroids(self.clusters)

            # check if clusters have changed
            if self._is_converged(centroids_old, self.centroids):
                self.isLast = True
                # If the position of each of the centers is compared with the old centers, the cycle is broken if the result is 0
                break

            if self.plot_steps:
                self.plot()

        return self._get_cluster_labels(self.clusters)

    def _get_cluster_labels(self, clusters):
        # Flowers that change sets are assigned to new labels
        labels = np.empty(self.n_samples)

        for cluster_idx, cluster in enumerate(clusters):
            for sample_index in cluster:
                labels[sample_index] = cluster_idx

        return labels

    def _create_clusters(self, centroids):
        # Assign the samples to the closest centroids to create clusters
        clusters = [[] for _ in range(self.K)]
        for idx, sample in enumerate(self.X):
            centroid_idx = self._closest_centroid(sample, centroids)
            clusters[centroid_idx].append(idx) # update flower's category
        return clusters

    def _closest_centroid(self, sample, centroids): # calculates where each flower is closer to the center and adds it to the appropriate cluster
        distances = [euclidean_distance(sample, point) for point in centroids]         # distance of the current sample to each centroid# calculate distance between flower and center
        closest_index = np.argmin(distances) # it takes closest centroid
        return closest_index

    def _get_centroids(self, clusters):
        centroids = np.zeros((self.K, self.n_features)) #takes the average of each of the centers according to the "sepal_length" and "sepal_width" of
        # its own flowers and their current positions are determined
        for cluster_idx, cluster in enumerate(clusters):
            cluster_mean = np.mean(self.X[cluster], axis=0)
            centroids[cluster_idx] = cluster_mean
        return centroids

    def _is_converged(self, centroids_old, centroids):
        # If the position of each of the centers is compared with the old centers, the cycle is broken if the result is 0
        distances = [
            euclidean_distance(centroids_old[i], centroids[i]) for i in range(self.K)
        ]
        print("distances" + str(distances))
        return sum(distances) == 0

    def plot(self):
        fig, ax = plt.subplots(figsize=(12, 8))
        

        for i, index in enumerate(self.clusters):
            point = self.X[index].T
            ax.scatter(*point)

        for point in self.centroids:
            ax.scatter(*point, marker="x", color="black", linewidth=2)
#QUESTION 5 ****************************************************************************************************************
        if (self.isLast):
            plt.clf()
            plt.subplot(1,2,1)
            plt.title("Last iteration")
            for i, index in enumerate(self.clusters):
                point = self.X[index].T
                plt.scatter(*point)
            for point in self.centroids:
                plt.scatter(*point, marker="x", color="black", linewidth=2)
            
            
            plt.subplot(1,2,2)
            plt.title("STEP 2")
            colors = {'Iris-setosa':'purple', 'Iris-virginica':'pink', 'Iris-versicolor':'orange'}
            labels = {'Iris-setosa': 'setosa', 'Iris-virginica': 'virginica','Iris-versicolor':'versicolor'}
            markers = {'Iris-setosa': 'v', 'Iris-virginica': '^','Iris-versicolor':'*'}
            flag1 = True
            flag2 = True
            flag3 = True
            for i in range(len(iris_sepalValues['sepal_length'])):
                if iris_species[i] =="Iris-setosa" and flag1 == True :
                    plt.scatter(iris_sepalValues['sepal_length'][i], iris_sepalValues['sepal_width'][i], s= 50 ,marker = markers[iris_species[i]], color=colors[iris_species[i]], label=labels[iris_species[i]])
                    flag1 = False

                elif iris_species[i] =="Iris-virginica" and flag2 == True :
                    plt.scatter(iris_sepalValues['sepal_length'][i], iris_sepalValues['sepal_width'][i] , s= 50, marker = markers[iris_species[i]], color=colors[iris_species[i]], label=labels[iris_species[i]])
                    flag2=False

                elif iris_species[i] == "Iris-versicolor" and flag3 == True:
                    plt.scatter(iris_sepalValues['sepal_length'][i], iris_sepalValues['sepal_width'][i], s= 50, marker = markers[iris_species[i]], color=colors[iris_species[i]], label=labels[iris_species[i]])
                    flag3 = False

                else:
                    plt.scatter(iris_sepalValues['sepal_length'][i], iris_sepalValues['sepal_width'][i], s= 50, marker = markers[iris_species[i]], color=colors[iris_species[i]])
            plt.legend()
            #Difference between last iteration and step 2 
# QUESTION 5 END ***********************************************************************************************************

# Testing
if __name__ == "__main__":
    from sklearn.datasets import make_blobs

    k = 3
    iris = sns.load_dataset('iris')
    X = iris.iloc[:, :2].values
    y = iris.iloc[:, 1].values

    clusters = 3

    k = KMeans(K=clusters, max_iters=150, plot_steps=True)
    y_pred = k.predict(X)

    k.plot()
```
